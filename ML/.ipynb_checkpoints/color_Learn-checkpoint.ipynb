```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm
import glob
```

```python
col_names = ["TopColor", "TrousersColor", "TopR", "TopG","TopB", "TrousersR", "TrousersG","TrousersB", "predict"]
df = pd.read_csv('./colors2.csv', sep=',', names=col_names)
```

```python
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.loc[:,  ["TopR", "TopG","TopB", "TrousersR", "TrousersG","TrousersB"]].values
y = df.loc[:, ['predict']].values
X_train, X_test, y_train, y_test = train_test_split(X ,y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
model = KNeighborsClassifier(n_neighbors=20,leaf_size=30)
model.fit(X,y)
#model.fit(X_train,y_train)
#from sklearn.metrics import accuracy_score
#accuracy_score(y_test,clf.predict(X_test))
```
